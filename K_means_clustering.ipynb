```python
netflix_movies = pd.read_csv('/content/netflix_movies')
```

```python
netflix_movies
```

```python
netflix_movies = netflix_movies.drop(netflix_movies.columns[0], axis=1)
```

```python
#netflix_movies['month'] = pd.to_datetime(netflix_movies['premiere']).dt.month
#netflix_movies['month']
```

```python
from sklearn.preprocessing import StandardScaler
X = netflix_movies.drop(columns=['title', 'genre', 'premiere', 'language',
'type', 'director', 'cast', 'description', 'features', 'isNetflix',
 'Similar Movie', 'movie_id', 'country', 'release_year', 'rating'])
#scaler = StandardScaler()
netflix_movies['Revenue'] = netflix_movies['Revenue'].replace(0, 1)
X['Revenue'] = np.log(netflix_movies['Revenue'])
#X = pd.get_dummies(X, columns=['rating', 'release_year'])
X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#scaler = StandardScaler()
#X = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(X)
clusters = kmeans.predict(X)
clusters
```

```python
netflix_movies['Cluster'] = clusters
netflix_movies
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
X_tsne
```

```python
netflix_movies['coordinates'] = [coordinates for coordinates in X_tsne]
netflix_movies
```

```python
coordinate_x = [coordinate[0] for coordinate in netflix_movies['coordinates']]
coordinate_y = [coordinate[1] for coordinate in netflix_movies['coordinates']]
visualization = pd.DataFrame()
visualization['x'] = coordinate_x
visualization['y'] = coordinate_y
visualization['title'] = netflix_movies['title']
visualization['cluster'] = netflix_movies['Cluster']
visualization
```

```python
visualization.to_csv('d3-scatterplot/mytext_new.tsv', sep='\t', index=False)
```

```python
mean_revenue = netflix_movies.groupby('Cluster')['Revenue'].mean()
#most common genre of movies in each cluster
grouped_genres = netflix_movies.groupby('Cluster').sum().iloc[:, 6:]
common_genre = grouped_genres.idxmax(axis=1)
#most common year
#grouped_years = netflix_movies.groupby('Cluster').sum().iloc[:, 64:72]
#common_year = grouped_years.idxmax(axis=1)
#most common rating
#grouped_ratings = netflix_movies.groupby('Cluster').sum().iloc[:, 72:]
#common_rating = grouped_ratings.idxmax(axis=1)
#mean runtime
mean_runtime = netflix_movies.groupby('Cluster')['runtime'].mean()
#mean imdb score
mean_imdb = netflix_movies.groupby('Cluster')['imdb score'].mean()
```

```python
netflix_movies_clusters = pd.DataFrame()
netflix_movies_clusters['cluster'] = np.arange(10)
netflix_movies_clusters['revenue'] = mean_revenue
netflix_movies_clusters['genre'] = common_genre
#netflix_movies_clusters['year'] = common_year
#netflix_movies_clusters['rating'] = common_rating
netflix_movies_clusters['runtime'] = mean_runtime
netflix_movies_clusters['imdb'] = mean_imdb
netflix_movies_clusters = netflix_movies_clusters.sort_values('revenue', ascending=False)
```
